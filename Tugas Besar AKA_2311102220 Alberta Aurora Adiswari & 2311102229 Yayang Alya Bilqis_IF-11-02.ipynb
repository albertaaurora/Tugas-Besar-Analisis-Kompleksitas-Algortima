```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable
```

```python
# Fungsi untuk menghitung total biaya dengan metode iteratif
def hitung_total_iteratif(jumlah_tiket, harga_tiket):
    total_biaya = 0
    for _ in range(jumlah_tiket):
        total_biaya += harga_tiket
    return total_biaya
```

```python
# Fungsi untuk menghitung total biaya dengan metode rekursif
def hitung_total_rekursif(jumlah_tiket, harga_tiket):
    if jumlah_tiket == 0:
        return 0
    else:
        return harga_tiket + hitung_total_rekursif(jumlah_tiket - 1, harga_tiket)
```

```python
# Grafik untuk menyimpan data
n_values = []
recursive_times = []
iterative_times = []
```

```python
# Fungsi untuk memperbarui grafik
def update_graph():
    plt.figure(figsize=(8, 6))
    plt.plot(n_values, recursive_times, label='Recursive', marker='o', linestyle='-')
    plt.plot(n_values, iterative_times, label='Iterative', marker='o', linestyle='-')
    plt.title('Performance Comparison: Recursive vs Iterative')
    plt.xlabel('Number of Tickets')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Fungsi untuk mencetak tabel waktu eksekusi
def print_execution_table():
    table = PrettyTable()
    table.field_names = ["Number of Tickets", "Recursive Time (s)", "Iterative Time (s)"]
    # Use the minimum length of all lists to avoid the IndexError
    min_len = min(len(n_values), len(recursive_times), len(iterative_times))
    for i in range(min_len):
        table.add_row([n_values[i], recursive_times[i], iterative_times[i]])
    print(table)
```

```python
# Program utama
while True:
    try:
        jumlah_tiket = int(input("Masukkan jumlah tiket yang ingin dibeli (atau ketik -1 untuk keluar): "))
        if jumlah_tiket == -1:
            print("Program selesai. Terima kasih!")
            break
        if jumlah_tiket < 0:
            print("Masukkan jumlah tiket yang valid!")
            continue

        kategori_tiket = input("Masukkan kategori tiket (VIP/Regular): ").lower()

        if kategori_tiket == 'vip':
            harga_tiket = 1000000  # Harga VIP
        else:
            harga_tiket = 500000  # Harga Regular

        n_values.append(jumlah_tiket)

        # Mengukur waktu eksekusi algoritma rekursif
        start_time = time.time()
        hitung_total_rekursif(jumlah_tiket, harga_tiket)
        recursive_times.append(time.time() - start_time)

        # Mengukur waktu eksekusi algoritma iteratif
        start_time = time.time()
        hitung_total_iteratif(jumlah_tiket, harga_tiket)
        iterative_times.append(time.time() - start_time)

        # Menampilkan tabel waktu eksekusi
        print_execution_table()

        # Memperbarui grafik
        update_graph()

    except ValueError:
        print("Masukkan jumlah tiket yang valid!")
```
